```python
class Atm:
  def __init__(self):
    self.pin = ''
    self.balance = 0
    self.menu()

  def menu(self):
    user_input= input('Welcome to the ATM: \n Enter 1 to create pin: \n Enter 2 to change pin : \n Enter 3 to check balance : \n Enter 4 to withdraw money : \n Enter 5 to exit.')
    if user_input =='1':
      self.create_pin()
    elif user_input == '2':
      self.change_pin()
    elif user_input == '3':
      self.check_balance()
    elif user_input == '4':
      self.withdraw()
    else:
      exit()

  def create_pin(self):
    user_pin = input('Enter your pin: ')
    self.pin = user_pin
    user_balance= input('Enter your balance : ')
    self.balance = user_balance
    print('Pin created successfully!!')
    self.menu()

  def change_pin(self):
    user_pin= input('Enter your old pin : ')
    if user_pin ==self.pin:
      new_pin=input('Enter your new pin :')
      self.pin= new_pin
      print('Pin changed successfully!!')
      self.menu()
    else:
      print('Invalid Pin')
      self.menu()

  def check_balance(self):
    user_pin = input('enter your pin : ')
    if user_pin == self.pin:
      print('Your balance is ', self.balance)
      self.menu()
    else:
      print('Invalid pin')
      self.menu()

  def withdraw(self):
    user_pin = input('Enter your pin : ')
    if user_pin == self.pin:
      amount = int(input('Enter an amount to withdraw :'))
      new_balance = int(self.balance) - amount
      print('Your total balance was : ', self.balance)
      self.balance = new_balance
      print(' Your new balance is ', self.balance)
      self.menu()

    else:
      print('Invalid Pin!!')
      self.menu()

obj1= Atm()


```

```python
class myData:
  def __init__(self,x,y):
    self.num=x
    self.den =y

  def __str__(self):
    return f'{self.num}/{self.den}'

  def __add__(self,data):
    new_num = self.num * data.den + self.den * data.num
    new_den = self.den * data.den
    return f'{new_num}/{new_den}'
```

```python
obj = myData(3,2)
obj2 = myData(4,5)
obj3 = obj + obj2
print(obj3)
```

```python
class Point:
  def __init__(self,x,y):
    self.x_cod = x
    self.y_cod = y

  def __str__(self):
    return f'<{self.x_cod},{self.y_cod}>'

  def euclidean_distance(self,other):
    return ((self.x_cod-other.x_cod)**2 - (self.y_cod-other.y_cod)**2)**0.5

  def distance_from_origin(self):
    return self.euclidean_distance(Point(0,0))


class Line:
  def __init__(self,A,B,C):
    self.A = A
    self.B = B
    self.C = C

  def __str__(self):
    return f'{self.A}x + {self.B}y + {self.C} = 0'

  def point_on_line(line, point):
    if line.A*point.x_cod + line.B*point.y_cod + line.C ==0:
      return ('lies on the line')
    else:
      return ('does not lie on the line')

  def shortest_distance(line,point):
    return abs((line.A*point.x_cod+ line.B * point.y_cod+ line.C)/((line.A)**2 + (line.B)**2)**0.5)


  def intersect(line1,line2):
    if line1.A*line2.B - line1.B*line2.A !=0:
      x = (line2.C*line1.B - line1.C*line2.B)/(line1.A*line2.B - line1.B*line2.A)
      y = (line1.C*line2.A - line2.C*line1.A)/(line1.A*line2.B - line1.B*line2.A)
      return Point(x,y)



```

```python
p1 = Point(1,12)
l1= Line(1,1,-2)
l2= Line(1,-1,2)
print(l1.point_on_line(p1))
# p2= Point(3,5)
print(p1.euclidean_distance(p2))
# print(p2.distance_from_origin())
# print(p1.distance_from_origin())
print(l1.shortest_distance(p1))
print(Line.intersect(l1,l2))

```

```python
# Accessing data using getter and setter

class Parent:
  def __init__(self,num):
    self.__num= num

  def getter(self):
    return self.__num


class Child(Parent):
  def show(self):
    print('This is the child class')

c = Child(100)
print(c.getter())
c.show()
```

```python

class Parent:
  def __init__(self,num):
    self.__num= num

  def getter(self):
    return self.__num


class Child(Parent):
  def __init__(self,val,num):
    self.val = val
    self.num = num
  def show(self):
    print('This is the child class')

c = Child(100)
print(c.getter())
c.show()
```

```python
# Super Constructor
# Super cannot access variable, cannot be used outside the class and it is used inside the child class.
class Vehicle:
  def __init__(self,price,type_of_veh):
    print('Constructor of parent class')
    self.price=price
    self.type_of_veh = type_of_veh

  def driving(self):
    print('i am driving it..')



class Scooter(Vehicle):
  def __init__(self,price,type_of_veh,model,fuel,mph):
    print('constructor of child class')
    super().__init__(price,type_of_veh)
    super().driving()
    self.model=model
    self.fuel= fuel
    self.mph=mph


s=Scooter(268000,'scootor','yamaha',5.0,70)
print(f'rs: {s.price}')
print(s.mph)

```

Polymorphism- is a concept in which same things act different in different situation
1. Method overriding: When two class has same method name, python will execute child method name
2. Method overloading: In a class, if there is same two method name, python will execute the lastest written method and may throw an error but, you can use provide default arguments if you want to run based on different input and it will provides different output.
3. Operator overloading: it is a concept in which same operator behaves different in different situation. For eg '+' concatenate in string, but adds in num and merge in list.
