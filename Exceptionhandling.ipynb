```python
try:
  with open('demo.txt','r') as f:
    print(f.read())
    print(5/0)
except ZeroDivisionError:
  print('Number cannot be divided by zero')
except NameError:
  print('name not defined')
except FileNotFoundError:
  print('file not found')
```

```python
# else: if your confindent about your try block code, you can write code in else block to execute it seemlessly
try:
  f=open('demo.txt','r')

except ZeroDivisionError:
  print('Number cannot be divided by zero')
except NameError:
  print('name not defined')
except FileNotFoundError:
  print('file not found')
else:
  print(f.read())
```

```python
# Finally: if code has error, it goes to exception and if code is correct, it goes to else and whatever is the case,
# it moves to finally statement. it means the code inside the finally block always get executed
try:
  f=open('demo.txt','r')

except ZeroDivisionError:
  print('Number cannot be divided by zero')
except NameError:
  print('name not defined')
except FileNotFoundError:
  print('file not found')
else:
  print(f.read())
finally:
  print('This will print anyhow no matter what.')
```

```python
# Raise Error: It will throw an error and Exception will catch the error.
# raise NameError("I don't know the name")

class Bank:
  def __init__(self,balance):
    self.balance=balance
  def withdraw(self,amount):
    if amount <0:
      raise Exception ('Amount cannot be negative.')
    if self.balance < amount:
      raise Exception("You don't have enough money")
    self.balance=self.balance-amount
    return self.balance

b=Bank(10000)
try:
  b.withdraw(2000)
except Exception as e:
  print(e)
else:
  print(b.balance)
```

```python
# Making custom exception: to take full control and application based error


class SecurityError(Exception):
  def __init__(self,messege):
    print(messege)

  def logout(self):
    print('logout')


class Person:
  def __init__(self,name,gmail,device):
    self.name=name
    self.gmail= gmail
    self.device = device


  def log_in(self,gmail,device):
    if device !=self.device:
      raise SecurityError('YOu are logged in from another device')
    if gmail == self.gmail and device==self.device:
      print('welcome to the club')
    else:
      print('Login error')




p =Person('ujjwal','ukharel@gmail.com','Andriod')

try:
  p.log_in('ukharel@gmail.com','Apple')
except SecurityError as e:
  e.logout()
else:
  print(p.name)
finally:
  print('Data connection closed...')
```
